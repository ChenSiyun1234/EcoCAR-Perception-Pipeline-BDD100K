# Stage3 / 03 — Compare stage3 vs stage1 best-row vs stage2 winner

Pulls the three `best_metrics.json` / `a5000_profile.json` files
and prints a 3-row comparison table.

```python
from google.colab import drive; drive.mount('/content/drive')
import os, sys, json, glob
REPO_ROOT = '/content/drive/MyDrive/EcoCAR/yolop_vehicle_lane'
os.chdir(REPO_ROOT); sys.path.insert(0, REPO_ROOT)
```

```python
def load_run(ckpt_dir):
    m = {}
    for name in ('best_metrics.json', 'a5000_profile.json'):
        p = os.path.join(ckpt_dir, name)
        if os.path.exists(p):
            with open(p) as f: m.update(json.load(f))
    return m

runs = {
    'stage1_best_row': '/content/drive/MyDrive/EcoCAR/yolop_vehicle_lane/stage1/checkpoints/yolopv2_best_row',
}
s2 = sorted(glob.glob('/content/drive/MyDrive/EcoCAR/yolop_vehicle_lane/stage2/checkpoints/*'))
if s2: runs['stage2_winner'] = s2[-1]
s3 = sorted(glob.glob('/content/drive/MyDrive/EcoCAR/yolop_vehicle_lane/stage3/checkpoints/*'))
if s3: runs['stage3_integrated'] = s3[-1]

for k, v in runs.items():
    print(k, '=>', v)
    print('   ', load_run(v))
```
